# Caminata aleatoria 1D

---

Una caminata aleatoria en una dimensión consiste en una partícula que se mueve en pasos discretos sobre una línea.

En cada paso, la partícula puede:

- Moverse a la derecha con probabilidad $p$
- Moverse a la izquierda con probabilidad $q$

donde:

$$
p + q = 1
$$

Sea $X_n$ la posición en el paso $n$, entonces:

$$
X_n =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } q
\end{cases}
$$

La posición después de $N$ pasos es:

$$
S_N = \sum_{i=1}^{N} X_i
$$

Esta simulación mostrará la trayectoria de la partícula.


---



## Código

```python
import random
import matplotlib.pyplot as plt
```

```python
# Probabilidad de ir a la derecha
p = 0.5

# numero de pasos
N = 50

# numero de caminatas
caminatas = 1
```

```python
for j in range(caminatas):

    posicion = 0
    trayectoria = [posicion]

    for i in range(N):

        q = random.random()

        if q < p:
            posicion = posicion + 1
        else:
            posicion = posicion - 1

        trayectoria.append(posicion)

    plt.plot(trayectoria)

plt.title("Caminata aleatoria 1D")
plt.xlabel("Paso")
plt.ylabel("Posición")
plt.grid(True)

plt.show()
```

# Análisis

---

Se puede observar que la partícula se mueve aleatoriamente a lo largo de la recta.

Cuando $p = 0.5$, no hay una dirección preferida, por lo que la caminata es simétrica.

Si se aumenta el valor de $p$, la partícula tiende a moverse más hacia la derecha.

Si se disminuye el valor de $p$, la partícula tiende a moverse hacia la izquierda.

Las trayectorias son diferentes cada vez debido a la naturaleza aleatoria del proceso.
